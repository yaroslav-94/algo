## A. Это кошка?

```python
def find_meow(s: str):
    s = s.lower()
    answer = [s[0]]
    ind = 1
    while ind<len(s):
        while ind<len(s) and s[ind]==answer[-1]:
            ind += 1
        if ind==len(s):
            break
        answer.append(s[ind])
    return 'YES' if ''.join(answer)=='meow' else 'NO'

t = int(input())
for _ in range(t):
    n = int(input())
    s = input()
    
    if len(s)<4:
        print('NO')
    else:
        print(find_meow(s))
    
```

## B. Посчитай количество пар

```python
def find_max_coast(n: int, k: int, s: str):
    d_alf = {}
    answer = 0
    
    for i in s:
        if i not in d_alf:
            d_alf[i] = 1
        else:
            d_alf[i] += 1
            
    alfabet = list(d_alf.keys())
    for j in range(len(alfabet)):
        for i in range(j+1, len(alfabet)):
            if alfabet[j].lower() == alfabet[i].lower():

                while min(d_alf[alfabet[j]], d_alf[alfabet[i]])>0:
                    d_alf[alfabet[j]] -= 1
                    d_alf[alfabet[i]] -= 1
                    answer += 1
                    
                while k>0 and d_alf[alfabet[j]]>1:
                    d_alf[alfabet[j]] -= 2
                    k -= 1
                    answer += 1
                    
                while k>0 and d_alf[alfabet[i]]>1:
                    d_alf[alfabet[i]] -= 2
                    k -= 1
                    answer += 1
                    
    for j in range(len(alfabet)):
        while k>0 and d_alf[alfabet[j]]>1:
            d_alf[alfabet[j]] -= 2
            k -= 1
            answer += 1     
                
    return answer

t = int(input())
for _ in range(t):
    n, k = list(map(int, input().split()))
    s = input()
    print(find_max_coast(n, k, s))
```

```python
import heapq

def find_max_power(arr: list, n: int):
    answer = 0
    bonus = []
    
    for i in range(n):
        if arr[i]>0:
            heapq.heappush(bonus, -arr[i])
            
        elif arr[i]==0 and len(bonus)>0:
            answer += -heapq.heappop(bonus)
            
    return answer

t = int(input())
for _ in range(t):
    n = int(input())
    arr = list(map(int, input().split()))
    print(find_max_power(arr, n))
```

```python
def count_val(n: int, arr: list):
    answer = 0
    d_arr = {}
    for i in range(n):
        d = {}
        for j in arr[i]:
            if j not in d:
                d[j] = 1
            else:
                d[j] += 1
        d_arr[i] = (
            set(d.keys()),
            sum(d.values()),
            d
        )
        
    for i in range(n):
        for j in range(i, n):
            fl = False
            d_new = d_arr[i][2]
            for key in d_arr[j][2].keys():
                if key not in d_new:
                    d_new[key] = d_arr[j][2][key]%2
                else:
                    d_new[key] = d_arr[j][2][key]%2+d_new[key]%2
            
            if (d_arr[i][1]+d_arr[j][1])%2==1 and len(d_arr[i][1] | d_arr[j][1])==25 and fl:
                answer += 1
        
    return answer


arr = [
    'ftl',
    'abcdefghijklmnopqrstuvwxy',
    'abcdeffghijkllmnopqrsttuvwxy',
    'ffftl',
    'aabbccddeeffgghhiijjkkllmmnnooppqqrrssttuuvvwwxxyy',
    'thedevid',
    'bcdefghhiiiijklmnopqrsuwxyz',
    'gorillasilverback',
    'abcdefg',
    'ijklmnopqrstuvwxyz',
]
assert count_val(n=10, arr=arr) == 5
```
