## Задача 1

Для разработки модуля отображения фотографий в виде вертикальной ленты решено использовать следующий алгоритм:

- выбрать k фотографий из имеющейся коллекции;
- отмасштабировать все фотографии до ширины w;
- отображать полученные фотографии в вертикальном формате одна над одной без отступов.
Определите, какую наибольшую и наименьшую высоту может иметь полученная лента из k фотографий.

Будем считать, что при масштабировании фотографии размера до ширины w новая высота вычисляется по следующей формуле:
$h=\frac{h_0*w}{w_0}$

Формат ввода
- первая строка $w:(640\leq w \leq 4096)$
- вторая строка $n,k:1 \leq k \leq n \leq 10^5$
- остальные n cтрок $w_i*h_i: (640 \leq w_i,h_i \leq 4096)$

Формат вывода
- Выведите в первой строке минимальную высоту ленты из k фотографий.
- Во второй строке выведите максимальную высоту ленты из k фотографий.

```python
# ширина
w = int(input())
n, k = list(map(lambda x: int(x), input().split()))
images = []

# ширина*высота на входе
for i in range(n):
    we, he = list(map(lambda x: int(x), input().split('x')))
    # преобразование к w
    if we!=w:
        he = he*w//we + 1 if (he*w/we)%1>0 else he*w//we
    images.append(he)
images.sort()

answer = 0
for i in range(k):
    answer += images[i]
print(answer)

answer = 0
for i in range(1, k+1):
    answer += images[-i]
print(answer)
    
```
